```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# Load the CSV file
uploaded_file = "Cr3.csv"
data = pd.read_csv(uploaded_file)

# Select numeric columns for various plots
numeric_columns = ['Temperature', 'Humidity', 'Ph', 'Rain Fall','Soil EC', 'Water TDS']

# Create subplots with 2 columns and 3 rows
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 20))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Units for each column
units = ['Â°C', '%', '', 'mm','dSm\u207B\u00B9', 'mgL\u207B\u00B9']

# Loop through numeric columns and create various plots
for i, column in enumerate(numeric_columns):
    # Boxplot with rotated x-axis labels
    sns.boxplot(x=data['Plant'], y=data[column], ax=axes[i], palette='pastel')
    axes[i].set_title(f'Boxplot - {column} by Plant')
    axes[i].set_xlabel('Plant', fontsize=14)
    if column == 'Ph':
        axes[i].set_xlabel('pH', fontsize=14)
    else:
        axes[i].set_xlabel(f'{column} ({units[i]})', fontsize=14)
    axes[i].tick_params(axis='both', which='both', labelsize=12)

    # Rotate x-axis labels
    axes[i].tick_params(axis='x', rotation=35)

# Adjust layout
plt.tight_layout()

# Save the plot with 400 dpi
file_path = "box_plots.png"
plt.savefig(file_path, dpi=400)

# Download the file from Colab
files.download(file_path)


```

**ANN model**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# Load the dataset
uploaded_file = "Cr3.csv"  # Replace with the actual file path
data = pd.read_csv(uploaded_file)

# Separate features and target variable
X = data.drop('Plant', axis=1)
y = data['Plant']

# Convert categorical labels to numerical values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build the ANN model for multiclass classification
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(len(label_encoder.classes_), activation='softmax')
])

# Compile the model for multiclass classification
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model and store training history
history = model.fit(X_train_scaled, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test)
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')

# Plot loss curve
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss', linewidth=4)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=4)
plt.title('Training and Validation Loss', fontsize=16)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.legend(fontsize=16)
plt.show()

# Plot accuracy curve
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy', linewidth=4)
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=4)
plt.title('Training and Validation Accuracy', fontsize=16)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.legend(fontsize=16)
plt.show()

# Display classification report
y_pred_probs = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = y_test
print('\nClassification Report:')
print(classification_report(y_true, y_pred, target_names=label_encoder.classes_))

# Plot confusion matrix
conf_matrix = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10, 8))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix', fontsize=16)
plt.colorbar()
tick_marks = np.arange(len(label_encoder.classes_))
plt.xticks(tick_marks, label_encoder.classes_, rotation=45)
plt.yticks(tick_marks, label_encoder.classes_)
plt.ylabel('True Label', fontsize=14)
plt.xlabel('Predicted Label', fontsize=14)
plt.show()

# Plot ROC curve
n_classes = len(label_encoder.classes_)
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test == i, y_pred_probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(10, 8))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'ROC Curve - Class {i} (AUC = {roc_auc[i]:.2f})', linewidth=4)

plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=16)
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.legend(fontsize=12)
plt.show()
```

**LSTM model**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# Load the dataset
uploaded_file = "Cr3.csv"  # Replace with the actual file path
data = pd.read_csv(uploaded_file)

# Separate features and target variable
X = data.drop('Plant', axis=1)
y = data['Plant']

# Convert categorical labels to numerical values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape data for LSTM input (assuming each sample is a sequence of length 1)
X_train_reshaped = np.reshape(X_train_scaled, (X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_reshaped = np.reshape(X_test_scaled, (X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

# Build the LSTM model
lstm_model = keras.Sequential([
    layers.LSTM(64, activation='relu', input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])),
    layers.Dense(len(label_encoder.classes_), activation='softmax')
])

# Compile the LSTM model
lstm_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the LSTM model and store training history
lstm_history = lstm_model.fit(X_train_reshaped, y_train, epochs=20, batch_size=32, validation_split=0.2)

# Evaluate the LSTM model on the test set
test_loss_lstm, test_accuracy_lstm = lstm_model.evaluate(X_test_reshaped, y_test)
print(f'Test Accuracy (LSTM): {test_accuracy_lstm * 100:.2f}%')

# Plot loss curve for LSTM
plt.figure(figsize=(12, 6))
plt.plot(lstm_history.history['loss'], label='Training Loss (LSTM)', linewidth=4)
plt.plot(lstm_history.history['val_loss'], label='Validation Loss (LSTM)', linewidth=4)
plt.title('Training and Validation Loss (LSTM)', fontsize=16)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.legend(fontsize=16)
plt.show()

# Plot accuracy curve for LSTM
plt.figure(figsize=(12, 6))
plt.plot(lstm_history.history['accuracy'], label='Training Accuracy (LSTM)', linewidth=4)
plt.plot(lstm_history.history['val_accuracy'], label='Validation Accuracy (LSTM)', linewidth=4)
plt.title('Training and Validation Accuracy (LSTM)', fontsize=16)
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.legend(fontsize=16)
plt.show()

# Display classification report for LSTM
y_pred_probs_lstm = lstm_model.predict(X_test_reshaped)
y_pred_lstm = np.argmax(y_pred_probs_lstm, axis=1)
y_true_lstm = y_test
print('\nClassification Report (LSTM):')
print(classification_report(y_true_lstm, y_pred_lstm, target_names=label_encoder.classes_))

# Plot confusion matrix for LSTM
conf_matrix_lstm = confusion_matrix(y_true_lstm, y_pred_lstm)
plt.figure(figsize=(10, 8))
plt.imshow(conf_matrix_lstm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix (LSTM)', fontsize=16)
plt.colorbar()
tick_marks_lstm = np.arange(len(label_encoder.classes_))
plt.xticks(tick_marks_lstm, label_encoder.classes_, rotation=45)
plt.yticks(tick_marks_lstm, label_encoder.classes_)
plt.ylabel('True Label', fontsize=14)
plt.xlabel('Predicted Label', fontsize=14)
plt.show()

# Plot ROC curve for LSTM
n_classes_lstm = len(label_encoder.classes_)
fpr_lstm = dict()
tpr_lstm = dict()
roc_auc_lstm = dict()

for i in range(n_classes_lstm):
    fpr_lstm[i], tpr_lstm[i], _ = roc_curve(y_test == i, y_pred_probs_lstm[:, i])
    roc_auc_lstm[i] = auc(fpr_lstm[i], tpr_lstm[i])

plt.figure(figsize=(10, 8))
for i in range(n_classes_lstm):
    plt.plot(fpr_lstm[i], tpr_lstm[i], label=f'ROC Curve (LSTM) - Class {i} (AUC = {roc_auc_lstm[i]:.2f})', linewidth=4)

plt.title('Receiver Operating Characteristic (ROC) Curve (LSTM)', fontsize=16)
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.legend(fontsize=12)
plt.show()
```
